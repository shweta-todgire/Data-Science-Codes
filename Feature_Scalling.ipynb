```python
## Standardizing a Feature
import numpy as np
import pandas as pd
from sklearn import preprocessing

df = pd.read_csv("StudentsPerformance.csv")

feature = df.select_dtypes(include=[np.number]).values

scaler = preprocessing.StandardScaler()
standardized = scaler.fit_transform(feature)

print("Original Feature Data:\n", feature)
print("\nStandardized Data:\n", standardized)
print("\nMean of standardized data:", standardized.mean())
print("Standard Deviation of standardized data:", standardized.std())
```

```python
## RobustScaler Method
import numpy as np
import pandas as pd
from sklearn import preprocessing

df = pd.read_csv("StudentsPerformance.csv")

feature = df.select_dtypes(include=[np.number]).values

robust_scaler = preprocessing.RobustScaler()
scaled_data = robust_scaler.fit_transform(feature)

print("Original Feature Data:\n", feature)
print("\nRobust Scaled Data:\n", scaled_data)
print("\nMedian of scaled data (approx):", np.median(scaled_data))
print("IQR of scaled data (approx):", np.subtract(*np.percentile(scaled_data, [75, 25])))
```

```python
## Normalizing Observations
import pandas as pd
from sklearn.preprocessing import Normalizer

df = pd.read_csv("StudentsPerformance.csv")

numeric_df = df.select_dtypes(include='number')
features = numeric_df.values

normalizer_l2 = Normalizer(norm="l2")
features_l2_norm = normalizer_l2.transform(features)
print("\nL2 Normalized Features:")
print(features_l2_norm)

normalizer_l1 = Normalizer(norm="l1")
features_l1_norm = normalizer_l1.transform(features)
print("\nL1 Normalized Features:")
print(features_l1_norm)

print("\nSum of first row (L1 Norm):", features_l1_norm[0].sum())
```

```python
## Grouping Observations Using Clustering
import pandas as pd
from sklearn.cluster import KMeans

df = pd.read_csv("StudentsPerformance.csv")

features = df.select_dtypes(include=['float64', 'int64']).values

clusterer = KMeans(n_clusters=3, random_state=0)
clusterer.fit(features)

df["group"] = clusterer.predict(features)

print(df.head())
```

```python
## Deleting Observations with Missing Values
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")

clean_df = df.dropna()

print(clean_df.head())
```

```python
## Imputing Missing Values
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

df = pd.read_csv("StudentsPerformance.csv")

features = df.select_dtypes(include=[np.number]).values

scaler = StandardScaler()
standardized_features = scaler.fit_transform(features)

true_value = standardized_features[0, 0]
standardized_features[0, 0] = np.nan

imputer = SimpleImputer(strategy="mean")
imputed_data = imputer.fit_transform(standardized_features)

print("True Value:", true_value)
print("Imputed Value:", imputed_data[0, 0])
```
