```python
import pandas as pd

df = pd.read_csv("/content/student_data.csv")
df.head()
```

```python
# One-sample t-test
# H0: mean studytime = 2, H1: mean studytime != 2

from scipy import stats

studytime = df['studytime']
t_stat, p_val = stats.ttest_1samp(studytime, 2)
print("One-sample t-test:")
print("t-statistic:", t_stat, "p-value:", p_val)
if p_val < 0.05:
    print("Reject H0: Average studytime is significantly different from 2.")
else:
    print("Fail to reject H0: No evidence that average studytime differs from 2.")
print("\n")
```

```python
# Two-sample t-test: Do males and females differ in final grade G3?
# H0: mean G3 (M) = mean G3 (F), H1: means differ

from scipy import stats

g3_male = df[df['sex']=='M']['G3']
g3_female = df[df['sex']=='F']['G3']
t_stat2, p_val2 = stats.ttest_ind(g3_male, g3_female)
print("Two-sample t-test:")
print("t-statistic:", t_stat2, "p-value:", p_val2)
if p_val2 < 0.05:
    print("Reject H0: Male and female final grades differ significantly.")
else:
    print("Fail to reject H0: No significant difference between male and female grades.")
print("\n")
```

```python
# Paired t-test: Improvement from G1 to G3
# H0: mean(G3 - G1) = 0, H1: mean(G3 - G1) != 0

from scipy import stats

t_stat3, p_val3 = stats.ttest_rel(df['G3'], df['G1'])
print("Paired t-test (G1 vs G3):")
print("t-statistic:", t_stat3, "p-value:", p_val3)
if p_val3 < 0.05:
    print("Reject H0: There is a significant improvement from G1 to G3.")
else:
    print("Fail to reject H0: No significant improvement from G1 to G3.")
print("\n")
```

```python
# Chi-square test: Are sex and internet access independent?
# H0: sex and internet are independent

from scipy import stats

contingency_table = pd.crosstab(df['sex'], df['internet'])
chi2, p_val4, dof, expected = stats.chi2_contingency(contingency_table)
print("Chi-square test:")
print("Chi2:", chi2, "p-value:", p_val4)
if p_val4 < 0.05:
    print("Reject H0: Sex and internet access are dependent.")
else:
    print("Fail to reject H0: Sex and internet access are independent.")
print("\n")
```

```python
# Z-test: Is average age different from 16? (assume known population SD=1.5)
# H0: mean age = 16

from statsmodels.stats.weightstats import ztest

age = df['age']
z_stat, p_val5 = ztest(age, value=16, ddof=1)
print("Z-test for age:")
print("z-statistic:", z_stat, "p-value:", p_val5)
if p_val5 < 0.05:
    print("Reject H0: Average age is significantly different from 16.")
else:
    print("Fail to reject H0: No evidence that average age differs from 16.")
```
