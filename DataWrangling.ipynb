```python
import pandas as pd
dataframe = pd.DataFrame()
```

```python
from google.colab import files

uploaded = files.upload()
df = pd.read_csv('StudentsPerformance.csv')

# Show first two rows
print(df.head(2))

# Show last two rows
print(df.tail(2))

# Show dimensions
print("Dimensions: {}".format(df.shape))

# Show statistics
print(df.describe())
```

```python
# Select the first row
print(df.iloc[0])

# Select three rows
print(df.iloc[:3])

# All rows up to and including the fourth row
print(df.iloc[:4])
```

```python
df_selected = df[df['math score'] > 90].head(2)
print(df_selected)

df_multiple = df[(df['math score'] > 70) & (df['reading score'] > 80)].head(2)
print(df_multiple)
```

```python
df['lunch'].replace("standard", "STD", inplace=True)
print(df.head(2))

df['lunch'].replace({"STD": "Standard", "free/reduced": "FR"}, inplace=True)
print(df.head(2))

df.replace("none", "None", inplace=True)
print(df.head(2))
```

```python
df.rename(columns={'math score': 'Math Score'}, inplace=True)
df.rename(columns={'reading score': 'Reading Score'}, inplace=True)
print(df.head(2))
```

```python
print("Maximum:", df['Math Score'].max())
print("Minimum:", df['Math Score'].min())
print("Mean:", df['Math Score'].mean())
print("Sum:", df['Math Score'].sum())
print("Count:", df['Math Score'].count())

# Dataframe Statistics
print(df.var(numeric_only=True))
print(df.std(numeric_only=True))
print(df.kurt(numeric_only=True))
print(df.skew(numeric_only=True))
```

```python
print(df.groupby('lunch').count())
```

```python
print(df['Math Score'].unique())
print(df['Math Score'].value_counts())
```

```python
print(df[df.isnull().any(axis=1)].head(2))
```

```python
df_dropped = df.drop('Math Score', axis=1)
print(df_dropped.head(2))
```

```python
df_filtered = df[df['Reading Score'] != 60]
print(df_filtered.head(2))

df_no_first = df.drop(df.index[0])
print(df_no_first.head(2))
```

```python
print(df.groupby('lunch').mean(numeric_only=True))
print(df.groupby('lunch')['Reading Score'].count())
```

```python
for val in df['lunch'][:2]:
    print(val)

#  List Comprehension
vals_upper = [str(val).upper() for val in df['lunch'][:2]]
print(vals_upper)
```

```python
def stringify(x):
    return str(x)

df['lunch_str'] = df['lunch'].apply(stringify)
print(df[['lunch', 'lunch_str']].head(2))
```

```python
print(df.groupby('lunch').count())
```
